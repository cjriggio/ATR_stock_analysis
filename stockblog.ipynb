```python
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates
import datetime
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pandas_datareader.data as web
```

```python
start = datetime.datetime(2019, 1, 1)
end = datetime.datetime(2019, 5, 31)
df_500 = web.DataReader('^GSPC', 'yahoo', start, end)
df_500.tail()
```

```python
def tr_to_dataframe(df):
    df.reset_index(inplace=True)
    for i in range(1, len(df)):
        Max, Min = max(df.loc[i, 'High'], df.loc[(i-1), 'Close']), min(df.loc[i, 'Low'], df.loc[(i-1), 'Close'])
        df.loc[i, 'TR'] = Max - Min
```

```python
tr_to_dataframe(df_500)
df_500.tail()
```

```python
df_500['Date'] = pd.to_datetime(df_500['Date'])
df_500["Date1"] = df_500["Date"].apply(mdates.date2num)
```

```python
ohlc= df_500[['Date1', 'Open', 'High', 'Low','Close']].copy()
plt.style.use('ggplot')

f1, ax = plt.subplots(figsize=(20,10))
candlestick_ohlc(ax, ohlc.values, width=.7, colorup='green', colordown='red');
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

#ax.plot(df_500['Date'], df_500['ATR_SMA_7'], color='blue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('S&P 500');
```

```python
atr_days = 3
i = atr_days
while i < len(df_500):
    df_500.loc[i, 'ATR_SMA_3'] = sum(list(df_500['TR'].loc[i-atr_days+1:i]))/atr_days
    i += 1
```

```python
atr_days = 7
i = atr_days
while i < len(df_500):
    df_500.loc[i, 'ATR_SMA_7'] = sum(list(df_500['TR'].loc[i-atr_days+1:i]))/atr_days
    i += 1
```

```python
atr_days = 14
i = atr_days
while i < len(df_500):
    df_500.loc[i, 'ATR_SMA_14'] = sum(list(df_500['TR'].loc[i-atr_days+1:i]))/atr_days
    i += 1
```

```python
df_500.drop('Date1', axis=1, inplace=True)
```

```python
df_500.tail(10)
```

```python
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(df_500['Date'], df_500['ATR_SMA_3'], color='blue')
ax.plot(df_500['Date'], df_500['ATR_SMA_7'], color='green')
ax.plot(df_500['Date'], df_500['ATR_SMA_14'], color='red')

plt.title('Average True Range')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend();
```
